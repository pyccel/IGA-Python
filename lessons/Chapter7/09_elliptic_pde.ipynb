#Â 9. Elliptic partial differential equation

In this tutorial, we consider the general elliptic problem

$$
\mathcal{L}(u) := - \nabla \cdot \left( A \nabla u \right) + \mathbf{b} \cdot \nabla u + c u
$$
where $A$, $\mathbf{b}$ and $c$ are constants.

we use the following weak formulation

$$
	a(v, u) = \int_{\Omega} A \nabla u \cdot \nabla v + \left( \mathbf{b} \cdot \nabla u \right) v + c u v
    ~d\Omega, \quad \forall u,v \in \mathcal{V}_h
$$
